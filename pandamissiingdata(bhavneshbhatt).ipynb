```python
import pandas as pd
import numpy as np
```

```python
data={'first_name':['john',np.nan,'sara','Donlad','sachin'],
      'last_name':['cena',np.nan,'ali','Teep','tendulkar'],
      'age':[41,np.nan,34,22,35],
      'gender':['m',np.nan,'f','m','m'],
      'preTestScore':[41,np.nan,np.nan,22,31],
      'postTestScore':[25,np.nan,np.nan,62,70],
      
    
    }
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### seaborn functions
## visualizing statistical relationships
### process of understanding realtionships b/w variables of a dataset  

```python
a = sns.load_dataset('flights')
sns.relplot(x='passengers',y='month',data=a)## 2 dimesnional
```

```python
a = sns.load_dataset('flights')## this is scatterd plot
sns.relplot(x='passengers',y='month',hue='year',data=a)## hue symentaic
```

```python
b=sns.load_dataset('tips')
sns.relplot(x='time',y='tip',data=b,kind='line')
```

### Multi plot grid
### graphs ara plotted side-by-side uisng the same scale and axes to aid comparsion to the programmers to differintate the information upto the laarge amount data
### facet or pair grid fucntion

```python
a=sns.load_dataset('iris')
b=sns.FacetGrid(a,col='species')
b.map(plt.hist,'sepal_length')
```

### pair grid
### pair of variabels

```python
a=sns.load_dataset('flights')
b=sns.PairGrid(a)
b.map(plt.scatter)
```

```python
## Color palettes
sns.set(style='darkgrid')
a=sns.load_dataset('flights')
b=sns.PairGrid(a)
b.map(plt.scatter)
```

```python
sns.set(style='dark')
a=sns.load_dataset('flights')
b=sns.PairGrid(a)
b.map(plt.scatter)
```

```python
sns.set(style='white',color_codes=True)
a=sns.load_dataset('tips')
sns.boxplot(x='day',y='total_bill',data=a)
```

```python
sns.set(style='white',color_codes=True)
a=sns.load_dataset('tips')
sns.boxplot(x='day',y='total_bill',data=a)
sns.despine(offset=10,trim=True)
```

### When should I use thr "inplace" parameter in pandas?

```python
ufo.shape
```

```python
ufo.head()
```

```python
ufo.drop('City',axis=1,inplace=False)
```

```python
ufo.head()
```

```python
ufo.drop('City',axis=1,inplace=True)
```

```python
ufo.head()
```

```python
#drop row if any value row is missing
ufo.dropna(how='any')
```

```python
ufo.dropna(how='any').shape
```

```python
ufo.shape
```

```python
ufo=ufo.set_index('Time')
```

```python
ufo.tail()
```

```python
ufo.head()
```

```python
ufo.fillna(method='bfill').tail()
```

```python
ufo.fillna(method='ffill').tail()
```

Basic Graph


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as nd
        
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

#Resize graph(dpi specifies pixels per inch.when saving probably should use 300 if possible)
#plt.figure(figsize=(5,3),dpi=300)

#Line 1
#Keyword argument notation
#plt.plot(x,y,label='2x',color='r',linestyle='--',linewidth=2,marker='.',markersize=10,markeredgecolor='blue')

#Short hand notation
#fm = [color][marker][line]

plt.plot(x,y,'b^--',label='2x')

##Line 2
x2 = np.arange(0,4.5,0.5)
#print(x2)

#plot part of the graph as line
plt.plot(x2[:6],x2[:6]**2,'r',label='x^2')

#plot remainder of graph as a dot
plt.plot(x2[5:],x2[5:]**2,'r--')

#Add a title(specify font parameters with fontdict)
plt.title('Our First Graph!',fontdict={'fontname':'Comic Sans MS','fontsize':20})

# X and Y Labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

#X , Y axis tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])

#Add a Legend
plt.legend()

#Save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('mygraph.png',dpi=300)

#show plot
plt.show()
```

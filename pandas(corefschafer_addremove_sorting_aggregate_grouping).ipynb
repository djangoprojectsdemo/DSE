```python
### what % of people from each country know python?
country_respondents =df['Country'].value_counts()
country_respondents
```

```python
## total number of people for each country who know python
country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x:x.str.contains('Python').sum())
country_uses_python
```

```python
###more than one series using concat function
python_df = pd.concat([country_respondents,country_uses_python],axis='columns',sort=False)
python_df
```

```python
python_df.rename(columns={'Country':'NumRespondents','LanguageWorkedWith':'NumKnowsPython'},inplace=True)
```

```python
python_df
```

```python
python_df['PectKnowsPython'] = (python_df['NumKnowsPython']/python_df['NumRespondents']) * 100
```

```python
python_df
```

```python
python_df.sort_values(by='PectKnowsPython',ascending=False)
python_df.head(50)
```

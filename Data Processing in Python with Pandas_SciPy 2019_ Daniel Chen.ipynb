```python
# R language (this piping) and python are also done piping
(billboard
 .melt(id_vars=['year','artist','track','time','date.entered'],
        value_name='rank',
        var_name='week')
 .groupby('artist')['rank']
 .mean()
)
```

```python
#another way notation in pandas
billboard\
 .melt(id_vars=['year','artist','track','time','date.entered'],
        value_name='rank',
        var_name='week')\
 .groupby('artist')['rank']\
 .mean()
```

### variables are stored in both rows and columns

```python
weather = pd.read_csv('weather_chen_daniel.csv')
```

```python
weather.head()
```

```python
weather_long = weather.melt(id_vars=['id','year','month','element'],
                           var_name ='day',
                           value_name ='temp')

```

```python
weather_long.head()
```

```python
#element column we want indivual columns(we do opposite of melt)
weather_long.pivot_table(index=['id','year','month','day'],
                        columns='element',
                        values ='temp')
```

```python
weather_long.pivot_table(index=['id','year','month','day'],
                        columns='element',
                        values ='temp').reset_index()
```

```python
(weather_long.pivot_table(index=['id','year','month','day'],
                        columns='element',
                        values ='temp')
 .reset_index()
)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips['tip'].plot(kind='hist')
plt.show()
```

```python
cts = tips['smoker'].value_counts()
```

```python
cts
```

```python
cts.plot(kind='bar')
plt.show()
```

```python
sns.countplot(x = 'smoker' ,data = tips)
```

```python
sns.distplot(tips['total_bill'])#distribution plot(continues variable columns)
#flexibly plot a univariate distribution of observations
```

```python
sns.distplot(tips['total_bill'],kde =False)
```

```python
#linear regression plot
sns.lmplot(x = 'total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x = 'total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x = 'total_bill',y='tip',data=tips,hue='sex',fit_reg=False)
```

```python
sns.lmplot(x = 'total_bill',y='tip',data=tips,hue='sex',fit_reg=False,col='smoker')
```

```python
sns.lmplot(x = 'total_bill',y='tip',data=tips,hue='sex',fit_reg=False,col='smoker',row ='day')
```

```python
facet = sns.FacetGrid(tips,col='time',row='smoker',hue='sex')
facet.map(plt.scatter,'total_bill','tip')
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

### Create a figure with 2 axes

### distplot of fare in one axes

### boxplot of class and fare on the other axes

```python
fig ,(ax1,ax2) = plt.subplots(1,2)
sns.distplot(titanic['fare'],ax =ax1)
sns.boxplot(x ='class',y='fare',data=titanic,ax=ax2)
```

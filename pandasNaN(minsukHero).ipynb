### How to check and replace NaN?

```python
import pandas as pd
```

```python
school_list =[{'name':'John','job':'Teacher','age':40},
              {'name':'Nate','job':'Teacher','age':35},
              {'name':'Yuna','job':'Teacher','age':37},
              {'name':'Abhram','job':'student','age':10},
              {'name':'Brain','job':'student','age':12},
              {'name':'gie','job':'student','age':11},
              {'name':'Nate','job':'teacher','age':None},
              {'name':'John','job':'student','age':None}
   
            ]
```

```python
df=pd.DataFrame(school_list,columns=['name','job','age'])
```

```python
df
```

```python
df.info()
```

```python
df.isna()
```

```python
df.isnull()
```

```python
df['age'] =df['age'].fillna(0)
```

```python
df
```

```python
school_list =[{'name':'John','job':'teacher','age':40},
              {'name':'Nate','job':'teacher','age':35},
              {'name':'Yuna','job':'teacher','age':37},
              {'name':'Abhram','job':'student','age':10},
              {'name':'Brain','job':'student','age':12},
              {'name':'gie','job':'student','age':11},
              {'name':'Nate','job':'teacher','age':None},
              {'name':'John','job':'student','age':None}
   
            ]
```

```python
df=pd.DataFrame(school_list,columns=['name','job','age'])
df
```

```python
df['age'].fillna(df.groupby('job')['age'].transform('median'),inplace=True)
```
